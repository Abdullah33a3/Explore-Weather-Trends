```python
import numpy as np
import matplotlib.pyplot as plt
import pwlf

def PWLR(input_1, output_1, n_break, Title):
    
    x = input_1
    y = output_1
    my_pwlf = pwlf.PiecewiseLinFit(x, y)
    breaks = my_pwlf.fit(n_break)
    print("Break Points:")
    print(breaks)
    x_hat = x
    y_hat = my_pwlf.predict(x)
    my_slopes = my_pwlf.slopes
    beta = my_pwlf.beta
    print("Slopes:")
    print(my_slopes)
    print("Intercepts:")
    intercepts_PWL = my_pwlf.intercepts
    print(intercepts_PWL)
    
    #Printing the equations for the lines
    i = 0
    while(i<len(my_slopes)):
        print("y = {}x + {}, for the range from {} up to {}".format(my_slopes[i], intercepts_PWL[i], breaks[i], breaks[i+1]))
        i = i+1
    
    plt.figure()
    plt.xlabel("Year",fontsize=size_of_font) 
    plt.ylabel("Average Temperature (C)",fontsize=size_of_font)
    plt.title(Title,fontsize=size_of_font);

    plt.plot(x, y, 'o')
    plt.plot(x_hat, y_hat, '-')
    Image_Name = "Piece_Wise_Regression.png"
    plt.savefig(Title + Image_Name)

    plt.show()

    return my_pwlf, my_pwlf.slopes
```

```python
input_01 = Global_Data_Reg.year.values.reshape(-1, 1).tolist()
output_01 = Global_Data_Reg.MV20_year_MA.values.reshape(-1, 1).tolist()
Flattened_Input_01 = [x for sublist in input_01 for x in sublist]
Flattened_Output_01 = [x for sublist in output_01 for x in sublist]
Title_Global = "Global Temperature"
model1, Slopes_01 = PWLR(Flattened_Input_01, Flattened_Output_01, 4, Title_Global)

input_02 = City_Data_Reg.year.values.reshape(-1, 1).tolist()
output_02 = City_Data_Reg.MV20_year_MA.values.reshape(-1, 1).tolist()
Flattened_Input_02 = [x for sublist in input_02 for x in sublist]
Flattened_Output_02 = [x for sublist in output_02 for x in sublist]
Title_Global = "Riyadh Temperature"
model2, Slopes_02 = PWLR(Flattened_Input_02, Flattened_Output_02, 3, Title_Global)
```
